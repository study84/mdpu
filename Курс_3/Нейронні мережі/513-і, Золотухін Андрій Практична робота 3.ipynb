```python
# Завдання 2 - Використання numpy
import numpy as np
```

```python
# 2.1. Створити масив numpy різних типів (одномірний, двомірний) та вивести їх розмірність.
from random import randint
rnd = lambda: randint(1, 100)

n1 = np.array([rnd() for _ in range(10)])
n2 = np.array([[rnd() for _ in range(10)], [rnd() for _ in range(10)]])
print(n1)
print(n2)
```

```python
# 2.2. Виконати базові арифметичні операції над масивами (додавання, множення).
n11 = np.array([rnd() for _ in range(10)])
n12 = np.array([rnd() for _ in range(10)])
print("N11", n11)
print("N12", n12)
print("N11 ** 2 =", n11 ** 2)     # піднесення до степеня
print("N11 + N12 =", n11 + n12)       # додавання елементів
print("N11 - N12 =", n11 - n12)       # віднімання
print("N11 * N12 =", n11 * n12)       # множення елементів
print("N11 / N12 =", n11 / n12)       # ділення елементів
```

```python
# 2.3. Створити масив значень, що розташовані в діапазоні від 0 до 1 з кроком 0.1.
arr = np.arange(0, 1.1, 0.1)   # додаємо 1.1, щоб включити 1.0
print(arr)
```

```python
# 2.4. Обчислити скалярний добуток двох одномірних масивів.
dot = np.dot(n11, n12)
print('скалярний добуток двох одномірних масивів', dot)
```

```python
# Завдання 3 - Використання matplotlib
import matplotlib.pyplot as plt
#print('matplotlib', matplotlib.__version__)
```

```python
# 3.1. Створити графік функції y = x^2 на проміжку від -10 до 10.
# 3.2. Додати заголовки до осей та заголовок графіку.
x = np.linspace(-10, 10, 200)   # 200 точок від -10 до 10
y = x ** 2                        # обчислюємо y = x²
# будуємо графік
plt.plot(x, y, label='y = x2', color='blue')
# оформлення
plt.title("Графік функції y = x2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
```

```python
# 3.3. Накласти на одному графіку декілька кривих (наприклад, y = x^2, y = x^3) з різними колірними мітками, додати легенду.
x = np.linspace(-10, 10, 200)
y1 = x ** 2
y2 = x ** 3
# будуємо графік
plt.plot(x, y1, label='y1 = x2', color='blue')
plt.plot(x, y2, label='y2 = x3', color='green')
# оформлення
plt.title("Графік функції y1 = x2, y2 = x3")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
```

```python
# 3.4. Створити гістограму за допомогою matplotlib на основі випадкового масиву значень.
data = np.random.randint(1, 101, 100)  # 100 чисел від 1 до 100

# будуємо гістограму
plt.hist(data, bins=10, color='skyblue', edgecolor='black')  # 10 бінів

# оформлення
plt.title("Гістограма випадкового масиву")
plt.xlabel("Значення")
plt.ylabel("Кількість")
plt.grid(axis='y', alpha=0.75)

# показати графік
plt.show()
```

```python
wine = load_wine()
# Завантаження вбудованого датасету
wine = load_wine()

# Створення DataFrame
df_wine = pd.DataFrame(data=wine.data, columns=wine.feature_names)

# Додаємо цільову змінну (якість вина)
df_wine['target'] = wine.target
```

```python
# 4.4. Виведіть інформацію про кількість записів у кожному класі (використайте метод .value_counts() на стовпці з мітками класів).
print("Кількість записів у кожному класі:")
print(df_wine['target'].value_counts())
```

```python
# Завдання 5 - Аналіз даних за допомогою scipy, matplotlib і pandas
# 5.1. Обчисліть середні значення для кожного з атрибутів в наборі даних (використовуйте numpy).
mean_values = np.mean(df_wine.iloc[:, :-1], axis=0)  # усі стовпці, крім target
print(mean_values)
```

```python
# 5.2. Побудова гістограм для кожного атрибуту
plt.figure(figsize=(12, 10))  # задаємо розмір фігури

for i, column in enumerate(df_wine.columns[:-1]):  # пропускаємо target
    plt.subplot(4, 4, i + 1)  # створюємо сітку 4x4
    plt.hist(df_wine[column], bins=15, color='skyblue', edgecolor='black')
    plt.title(column, fontsize=9)
    plt.tight_layout()  # щоб графіки не накладалися

plt.suptitle('Гістограми атрибутів набору Wine', fontsize=14, y=1.02)
plt.show()
```

```python
# 5.3. Побудова boxplot для кожного атрибуту
plt.figure(figsize=(12, 8))

# Використовуємо метод pandas .boxplot()
df_wine.iloc[:, :-1].boxplot(rot=45, grid=True, patch_artist=True)

plt.title('Бокс-плоти для атрибутів набору Wine', fontsize=14)
plt.ylabel('Значення')
plt.xlabel('Атрибути')
plt.tight_layout()
plt.show()
```

```python
# 5.4. Бокс-плоти з розділенням за класами:
plt.figure(figsize=(15, 10))

for i, column in enumerate(df_wine.columns[:-1]):  # пропускаємо target
    plt.subplot(4, 4, i + 1)
    df_wine.boxplot(column=column, by='target', grid=False)
    plt.title(column)
    plt.suptitle("")  # вимикаємо заголовок, який pandas додає автоматично
    plt.xlabel("Клас вина")
    plt.ylabel("Значення")

plt.tight_layout()
plt.suptitle("Порівняння розподілу атрибутів між класами", fontsize=16, y=1.02)
plt.show()
```

```python
# 5.5. Обчислення кореляційної матриці
import seaborn as sns
corr_matrix = df_wine.iloc[:, :-1].corr()  # пропускаємо 'target'

print("Кореляційна матриця атрибутів:\n")
print(corr_matrix)

# (додатково) Візуалізація тепловою картою
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Кореляційна матриця атрибутів (Wine Dataset)")
plt.show()
```
